```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/Dataset/dat5.csv"
df=pd.read_csv(path)
df.columns =['Deck1','Result','Deck2','Dice Roll']
```

```python
#Check the overall winrate of deck1

result_counts = df['Result'].value_counts()
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%',shadow=True)
plt.title('Overall Winrate Of Deck1')
plt.show()
```

```python
#Change the value of deck to see the match history
#And winrate between deck1 and the chosen deck2
deck = "Lab"
deck_df = df[df['Deck2'] == deck]
print("Match History vs",{deck}, "\n", deck_df)

record = deck_df['Result'].value_counts()
print("Record against ",{deck},":\n", record)

plt.pie(record, labels=record.index, autopct='%1.1f%%', shadow=True)
plt.title(f'Win% vs {deck}')
plt.show()
```

```python
#check and display the winrate of deck1 when it wins the dieroll
deck1="Snake-eyes"
deck1_df = df[df['Dice Roll'] == deck1]
dice_record_win = deck1_df['Result'].value_counts()

plt.pie(dice_record_win, labels=dice_record_win.index, autopct='%1.1f%%', shadow=True)
plt.title('Win% when wins the die roll')
plt.show()
```

```python
#check and display the winrate of deck1 when it loses the dieroll
deck2="Snake-eyes"
deck2_df = df[df['Dice Roll'] != deck2]
dice_record_lost = deck2_df['Result'].value_counts()

plt.pie(dice_record_lost, labels=result_counts.index, autopct='%1.1f%%', shadow=True)
plt.title('Win% when loses the die roll')
plt.show()
```

```python
#calculates and displays the highest and lowest winrate decks vs deck1

deck_record = df.groupby('Deck2')['Result'].value_counts().unstack(fill_value=0)

deck_record['WinRate'] = deck_record['Win'] / (deck_record['Win'] + deck_record['Lose'])

lowest_win_rate_deck = deck_record['WinRate'].idxmin()
lowest_win_rate = deck_record['WinRate'].min()

highest_win_rate_deck = deck_record['WinRate'].idxmax()
highest_win_rate = deck_record['WinRate'].max()

print(f"The worst matchup is {lowest_win_rate_deck} win% {lowest_win_rate:.2%}")
print(f"The best matchup is {highest_win_rate_deck} win% {highest_win_rate:.2%}")
```
